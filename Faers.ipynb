```python
import pandas as pd

# Define the paths to your data files
demo_path = '/content/drive/MyDrive/Research_notebooks/FAERS/DEMO24Q2.txt'
drug_path = '/content/drive/MyDrive/Research_notebooks/FAERS/DRUG24Q2.txt'
reac_path = '/content/drive/MyDrive/Research_notebooks/FAERS/REAC24Q2.txt'
outc_path = '/content/drive/MyDrive/Research_notebooks/FAERS/OUTC24Q2.txt'
rpsr_path = '/content/drive/MyDrive/Research_notebooks/FAERS/RPSR24Q2.txt'
ther_path = '/content/drive/MyDrive/Research_notebooks/FAERS/THER24Q2.txt'
indi_path = '/content/drive/MyDrive/Research_notebooks/FAERS/INDI24Q2.txt'

# Function to count unique case IDs
def count_unique_caseids(file_path, delimiter='$'):
    df = pd.read_csv(file_path, delimiter=delimiter, usecols=['caseid'])
    return df['caseid'].nunique()

# Count unique case IDs in each file
demo_caseids = count_unique_caseids(demo_path)
drug_caseids = count_unique_caseids(drug_path)
reac_caseids = count_unique_caseids(reac_path)
outc_caseids = count_unique_caseids(outc_path)
rpsr_caseids = count_unique_caseids(rpsr_path)
ther_caseids = count_unique_caseids(ther_path)
indi_caseids = count_unique_caseids(indi_path)

# Print the counts
print(f"Unique case IDs in DEMO24Q2.txt: {demo_caseids}")
print(f"Unique case IDs in DRUG24Q2.txt: {drug_caseids}")
print(f"Unique case IDs in REAC24Q2.txt: {reac_caseids}")
print(f"Unique case IDs in OUTC24Q2.txt: {outc_caseids}")
print(f"Unique case IDs in RPSR24Q2.txt: {rpsr_caseids}")
print(f"Unique case IDs in THER24Q2.txt: {ther_caseids}")
print(f"Unique case IDs in INDI24Q2.txt: {indi_caseids}")
```

```python
import pandas as pd
import os

# Define the paths to your data files
demo_path = '/content/drive/MyDrive/Research_notebooks/FAERS/DEMO24Q2.txt'
drug_path = '/content/drive/MyDrive/Research_notebooks/FAERS/DRUG24Q2.txt'
reac_path = '/content/drive/MyDrive/Research_notebooks/FAERS/REAC24Q2.txt'
outc_path = '/content/drive/MyDrive/Research_notebooks/FAERS/OUTC24Q2.txt'
rpsr_path = '/content/drive/MyDrive/Research_notebooks/FAERS/RPSR24Q2.txt'
ther_path = '/content/drive/MyDrive/Research_notebooks/FAERS/THER24Q2.txt'
indi_path = '/content/drive/MyDrive/Research_notebooks/FAERS/INDI24Q2.txt'

# Load the DRUG file and filter for Lorlatinib
drug_df = pd.read_csv(drug_path, delimiter='$')
lorlatinib_df = drug_df[drug_df['drugname'].str.contains('Lorlatinib', case=False, na=False)]

# Get the unique primaryid and caseid for Lorlatinib
lorlatinib_ids = lorlatinib_df[['primaryid', 'caseid']].drop_duplicates()

# Function to load and filter data based on primaryid and caseid
def load_and_filter(file_path, ids_df, delimiter='$'):
    df = pd.read_csv(file_path, delimiter=delimiter)
    filtered_df = df.merge(ids_df, on=['primaryid', 'caseid'], how='inner')
    return filtered_df

# Load and filter each file
demo_df = load_and_filter(demo_path, lorlatinib_ids)
reac_df = load_and_filter(reac_path, lorlatinib_ids)
outc_df = load_and_filter(outc_path, lorlatinib_ids)
rpsr_df = load_and_filter(rpsr_path, lorlatinib_ids)
ther_df = load_and_filter(ther_path, lorlatinib_ids)
indi_df = load_and_filter(indi_path, lorlatinib_ids)

# Aggregate the REAC data to avoid duplicate primaryid and caseid
reac_agg = reac_df.groupby(['primaryid', 'caseid'])['pt'].apply(lambda x: '; '.join(x)).reset_index()

# Merge the DataFrames on 'primaryid' and 'caseid'
merged_df = demo_df.merge(lorlatinib_df, on=['primaryid', 'caseid'], how='left') \
                   .merge(reac_agg, on=['primaryid', 'caseid'], how='left') \
                   .merge(outc_df, on=['primaryid', 'caseid'], how='left') \
                   .merge(rpsr_df, on=['primaryid', 'caseid'], how='left') \
                   .merge(ther_df, on=['primaryid', 'caseid'], how='left') \
                   .merge(indi_df, on=['primaryid', 'caseid'], how='left')

# Save the merged DataFrame to an Excel file
output_path = '/content/drive/MyDrive/Research_notebooks/FAERS/Lorlatinib_Consolidated_2024Q2_1.xlsx'
merged_df.to_excel(output_path, index=False)

print(f"Consolidated data for Lorlatinib saved to {output_path}")
```
